```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import sys
import os
```

```python
# I will use the wine quality dataset
X = load_breast_cancer().data
y = load_breast_cancer().target

y = np.where(y == 0, -1, 1)

print(X.shape)
print(y.shape)

print(np.unique(y, return_counts = True ))

"""
There are 569 instances of data with 212 instances of class 0 and 357 instances of class 1.
There are 30 features.
"""
instances, features = X.shape

X_train , X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
class SVM:
  def __init__(self,
               learning_rate=0.001,
               lambda_param=0.01,
               n_iters=1000):

    # "Each class should have a constructor which initializes the weights and bias,"
    # ->> to initialize the weights and biases, the dimnensions should be known

    #instances, features = X.shape


    self.lr = learning_rate
    self.lambda_param = lambda_param

    self.n_iters = n_iters
    self.w = None
    self.b = None


  def forward(self, X_):

    print("inside forward function:\n X shape : ", X_.shape)
    print("X transpose shape : ", X_.transpose().shape)

    print("self.w shape : ", self.w.shape)
    print("self.b shape : ", self.b.shape)
    print("returning... " , ( X_.transpose() @ self.w) - self.b)
    #a function forward which computes the output of the ML model based on a given input
    return ( X_.transpose() @ self.w).item() - self.b

  #def hinge_loss(self, y, y_pred):
  #  return np.maximum( self.lambda_param * np.linalg.norm(self.w, ord=2) , self.lambda_param * np.linalg.norm(self.w, ord=2) + (1 - y * y_pred) )


  #def partial_derivative(self, y , y_pred)
  #  k = y.shape[0]


  def fit(  self, X, y):

    #print("---- in fit function")

    n_samples, n_features = X.shape

    #print("n_samples, n_features", n_samples, n_features)

    self.w = np.random.rand(n_features, 1)
    self.b = np.random.rand(1)


    print("self.w shape : ", self.w.shape)
    print("self.b shape : ", self.b.shape)

    y = np.where(y <= 0, -1, 1)

    #sys.exit(0)

    for _ in range(self.n_iters):
      for idx, x_i in enumerate(X): #


        x_i = x_i.reshape((-1, 1))
        print("--- in fit : x_i shape : ", x_i.shape)

        y_i_pred = self.forward(x_i)

        print("y_i_pred : ", y_i_pred)
        print("y_i_pred.shape" , y_i_pred.shape)
        #sys.exit(0)
        print(y_i_pred.shape)
        print(y[idx].shape)



        if (y[idx] * y_i_pred >= 1):
          self.w -= self.lr * (2 * self.lambda_param * self.w) # updating weight by a factor of regularization term
          self.b -= 0 # not updating the bias

        else: # penalize  the weight and bias  for making error wrt to the weight
          self.w -= self.lr * ( 2 * self.lambda_param * self.w - (x_i * y[idx]) ) # no bias was used in this eqn
          self.b -= self.lr * y[idx]

  def predict(self, X):
        return np.sign(np.dot(X, self.w) - self.b)
```

```python
# Set random seed for reproducibility
np.random.seed(0)

# Generate 6 samples for label +1
X_pos = np.random.rand(6, 3)  # 6 samples, 3 features
y_pos = np.ones(6)  # Label +1

# Generate 6 samples for label -1
X_neg = np.random.rand(6, 3)  # 6 samples, 3 features
y_neg = -np.ones(6)  # Label -1

# Combine positive and negative samples
X = np.vstack((X_pos, X_neg))  # Shape: (12, 3)
y = np.concatenate((y_pos, y_neg))  # Shape: (12,)

# Output the generated dataset
print("Input Features (X):")
print(X)
print("\nLabels (y):")
print(y)
```

```python
# Initialize the SVM model
model = SVM(learning_rate=0.01, n_iters=10)

# Fit the model on the dataset
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Print results
print("Predictions:")
print(y_pred.flatten())
print("\nTrue Labels:")
print(y_test)
print("accuracy : ", accuracy_score(y_test, y_pred))
```
